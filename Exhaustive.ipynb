# 貪欲アルゴリズム

```python
import itertools
import math
import random
import numpy as np
```

```python
n = 20
l = list(range(0,n))
w = [random.randint(0,50) for i in range(n)] 
W = random.randint(0,100)
# l = list(range(0,6))
# w = [4,7,3,2,6,5]
# W = 9
print(w)
print(W)
```

```python
S = []
R = []
MW = 0
M = []

for i in range(len(l)+1):
    for v in itertools.combinations(l,i):
        for j in range(len(v)):
            R.append(w[v[j]])
        if sum(R) == W:
            if MW != W:
                MW = W
                S = []
                S.append(v)
            elif MW == W:
                S.append(v)
        elif sum(R) < W:
            if MW < sum(R):
                MW = sum(R)
                S = []
                S.append(v)
            elif MW == sum(R):
                S.append(v)
        R = []
print(S)
```

# DP法

```python
M = np.empty((len(w)+1, W+1))
for j in range(W+1):
    M[0][j] = 0
for i in range(len(w)):
    for j in range(W+1):
        if j < w[i]:
            M[i+1][j] = M[i][j]
        else:
            M[i+1][j] = max(M[i][j], w[i] + M[i][j-w[i]])
            
print(M)
#print(V)
```

```python
def MakeCombination(w,V,i,j):
    if V[i][j-w[i]] == ['*']:
        V[i+1][j].append((i,))
    else:
        for k in range(len(V[i][j-w[i]])):
            V[i+1][j].append(V[i][j-w[i]][k]+(i,))

V = [[[] for j in range(W+1)] for i in range(len(w)+1)]
for i in range (W+1):
    V[0][i].append('*')

for i in range(len(w)):
    for j in range(W+1):
        if j < w[i]:
            V[i+1][j] = V[i][j]
        elif M[i][j] > w[i] + M[i][j-w[i]]:
            V[i+1][j] = V[i][j]
        elif M[i][j] < w[i] + M[i][j-w[i]]:
            MakeCombination(w,V,i,j)
        elif M[i][j] == w[i] + M[i][j-w[i]]:
            MakeCombination(w,V,i,j)
            for k in range(len(V[i][j])):
                V[i+1][j].append(V[i][j][k])
print(V[len(w)][W])
            
```
